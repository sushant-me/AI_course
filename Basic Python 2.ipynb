```python
# Simple Calculator
```

```python
def add(a, b):
    return f"Addition of {a} and {b} is: {a + b}"

def sub(a, b):
    return f"Subtract of {a} and {b} is: {a - b}"

def mul(a, b):
    return f"Multiplication of {a} and {b} is: {a * b}"

def div(a, b):
    return f"Division of {a} and {b} is: {a / b}"
```

```python
def calculator(): # main function
    print("1. Add")
    print("2. Subtract")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")

    user_choice = int(input("Enter number from 1 - 5: "))

    if user_choice == 5:
        print("Thank you for choosing us..")
        return

    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    if user_choice == 1:
        print(add(num1, num2))
    elif user_choice == 2:
        print(sub(num1, num2))
    elif user_choice == 3:
        print(mul(num1, num2))
    elif user_choice == 4:
        print(div(num1, num2))
    else:
        print("Invalid Input. Choose Number from 1 -  5.")
```

```python
calculator()
```

# *Args (All Arguments)

```python
def person(name, *details):
    print("Name of person is:",name)
    for each_detail in details:
        print(each_detail)
```

```python
person("Ram",6787839330, 30, "Kathmandu", "Bachelors")
```

```python
# **kwargs (All Keyword Arguments)
def person(name, **details):
    print("Name of person is:",name)
    for key, each_detail in details.items():
        print(f"{key} of person is: {each_detail}")
```

```python
person("Ram", age=30, phone=9866735635, address="Kathmandu", qualification="Bachelors")
```

## Inheritance

```python
class Animal:
    def sound(self):
        print("Animals make sound!!")
# Inherit Animal
class Dog(Animal):
    pass
```

```python
d = Dog()
d.sound()
```

```python
class Animal:
    def sound(self):
        print("Animals make sound!!")
# Inherit Animal
class Dog(Animal):
    def __init__(self, name, color):
        self.name = name
        self.color = color
```

```python
dog = Dog("Charlie", "Golden")
print(dog.name)
print(dog.color)
dog.sound()
```

```python
# Encapsulation
# public
# protected (_)
# private (__)
```

```python
class BankAccount():
    def __init__(self, name, balance):
        self.name = name # Public
        self.__balance = balance # Private

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance
```

```python
acc = BankAccount("Shyam", 4000)
print("Current Balance:",acc.get_balance())
# deposit
acc.deposit(3000)

print("Balance after deposit:",acc.get_balance())
```

# Polymorphism

```python
class CreditCard:
    def pay(self, amount):
        return f"{amount} paid through credit card!!"

class DebitCard:
    def pay(self, amount):
        return f"{amount} paid through debit card!!"

class Cash:
    def pay(self, amount):
        return f"{amount} paid through cash!!"

def make_payment(payment_method, amount):
    print(payment_method.pay(amount))
```

```python
make_payment(Cash(), 300)
```

```python
class PaymentMethod: # parent class
    def make_payment(self, amount):
        return "Make payment through Debit Card, Credit Card or Cash!!"
        
class DebitCard(PaymentMethod):
    def make_payment(self, amount):
        return f"Paid Rs. {amount} using Debit Card!!"

class CreditCard(PaymentMethod):
    def make_payment(self, amount):
        return f"Paid Rs. {amount} using Credit Card!!"

class Cash(PaymentMethod):
    def make_payment(self, amount):
        return f"Paid Rs. {amount} using Cash!!"

def payment(payment_method, amount):
    print(payment_method.make_payment(amount))
```

```python
dc = DebitCard()
cc = CreditCard()
c = Cash()
```

```python
payment(dc, 1200)
payment(cc, 11000)
payment(c, 100000)
```

# Abstraction

```python
from abc import ABC, abstractmethod # Abstract Base Class
```

```python
class Payment(ABC): # parent class
    @abstractmethod
    def pay(self, amount):
        pass

    @abstractmethod
    def connection(self):
        print("Connection successfull!!")

class CreditCard(Payment): # child class
    def pay(self, amount):
        return f"{amount} paid through credit card!!"

    def connection(self):
        return "Connection Successfull!!"
```

```python
c = CreditCard()
print(c.pay(3000))
print(c.connection())
```
