# Dictionary

```python
# Key and Value pair
# {}
data = {
    "name": ['Ram', 'Shyam', 'Hari'],
    "age": [35, 20, 55],
    "city": ['Lalitpur', 'Kathmandu', 'Bhaktapur'],
    "Country": ["France", "USA", "UK"]
}
print(data)
```

```python
for key in data.keys():
    print(key)
```

```python
for value in data.values():
    print(value)
```

```python
for key, value in data.items():
    print(f"{key}: {value}")
```
